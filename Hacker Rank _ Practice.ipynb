```python
n = int(input())
if(n in range(2, 11)):
    student_marks = {}
    for _ in range(n):
        name, *line = input().split()
        for i in line:
            i = float(i)
            
            if(i < 0 and i > 100):
                print(i)
                print('Please enter a valid value')
                break
                
        if(len(line) == 3):
            scores = list(map(float, line))
            student_marks[name] = scores
        else:
            print('The length of marks array should be 3')
            break
    query_name = input()
    if query_name in student_marks.keys():
        l1 = student_marks[query_name]
        avg = sum(l1)/len(l1)
        print('{0:.2f}'.format(avg))
    else:
        print('Name doesnot exist')
else:
    print('Please enter a valid value')
```

```python
n = int(input())
op_list = []
for _ in range(n):
    command, *values = input().split()
    values = list(map(int, values))
    if command == 'print':
        print(op_list)
    elif command == 'insert':
        op_list.insert(values[0], values[1])
    elif command == 'remove':
        op_list.remove(values[0])
    elif command == 'append':
        op_list.append(values[0])
    elif command == 'sort':
        op_list.sort()
    elif command == 'pop':
        op_list.pop()
    elif command == 'reverse':
        op_list.reverse()
    else:
        print('Insert a valid command with valid inputs')
    
    
```

```python
def count_substring(string, sub_string):
    if(len(string) in range(1, 201)):
        counter = 0
        i = 0
        length_str = len(string)

        while i < length_str:
            print(i)
            sub_str_value = string.find(sub_string, i, length_str-1)
            if(sub_str_value >= 0):
                counter += 1
                continue
            else:
                break
            i =  i + sub_str_value
            
    return counter

if __name__ == '__main__':
    string = input().strip()
    sub_string = input().strip()
    
    count = count_substring(string, sub_string)
    print(count)
```

```python
def print_rangoli(size):
    import string
    alpha = string.ascii_lowercase

    L = []
    for i in range(size):
        s = "-".join(alpha[i:size])
        print(s[::-1] + s[1:])
        L.append((s[::-1]+s[1:]).center(4*size-3, "-"))
        
    print('\n'.join(L[:0:-1]+L))

if __name__ == '__main__':
    n = int(input())
    print_rangoli(n)
```
